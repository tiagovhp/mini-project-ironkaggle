```python
import pickle
import pandas as pd
import numpy as np
```

# Import the Machine Learning that predict Sales

```python
#Import our model
model = pickle.load(open("trained_model.pkl", "rb"))
```

# Import the real data

```python
# Import the real data
real_data = pd.read_csv("REAL_DATA.csv")
```

```python
# Let's check the data

real_data.head()
```

# Prepare the real data for the model

```python
# Prepare the real data for predictions
data_for_model = real_data

data_for_model

```

```python
## 1. 'state_holiday' -> one hot encode to state_holiday_a, state_holiday_b, state_holiday_c
data_for_model['state_holiday'].unique()
data_for_model = pd.get_dummies(data_for_model, columns=['state_holiday'], prefix = 'state_holiday', drop_first = True)

## 2. date
data_for_model['date'] = pd.to_datetime(data_for_model['date'], format='%d/%m/%Y').dt.strftime('%Y-%m-%d')

data_for_model

### convert into integer
data_for_model['date'] = data_for_model['date'].str.replace("-" , "").astype("int64")
```

```python
## 3. 'day_of_week' -> one hot encoding
data_for_model = pd.get_dummies(data_for_model, columns=['day_of_week'], prefix = 'day_of_week', drop_first = False)
data_for_model.head()
```

```python
data_for_model = data_for_model.drop(columns='index')
data_for_model
```

```python
# Standardize the data
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
data_for_model = pd.DataFrame(scaler.fit_transform(data_for_model), columns=data_for_model.columns)
data_for_model
```

# Make predictions with real data

```python
# Make predictions with real data

pred = model.predict(data_for_model)
pred = np.clip(pred,a_min=0,a_max=None)
pd.DataFrame(pred)
```

```python
real_data['sales']=pred

real_data

```

# Export the real data with predictions

```python
real_data.to_csv('G2_2.csv', index=True)
```
